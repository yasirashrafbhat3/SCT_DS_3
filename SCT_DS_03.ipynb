```python
column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 -1.5 * iqr
upper_bound = q3 + 1.5* iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
```

```python
df1 = df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols, inplace=True, axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['y'].value_counts()
```

```python
x = df_encoded.drop('y', axis= 1)
z =df_encoded['y']
print(x.shape)
print(z.shape)
print (type(x)) 
print (type(z))
```

```python
from sklearn.model_selection import train_test_split

print(4119*0.25)
```

```python
x_train, x_test, z_train, z_test = train_test_split(x,z,test_size=0.25, random_state=1)
print(x_train.shape)
print(x_test.shape)
print(z_train.shape)
print(z_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score 

def eval_model(z_test,z_pred): 
    acc = accuracy_score(z_test,z_pred)
    print ('Accuracy_Score', acc)
    cm = confusion_matrix(z_test,z_pred)
    print ('Confusion Matrix\n',cm)
    print( 'Classification Report\n', classification_report(z_test,z_pred))

def mscore(model):
    train_score = model.score(x_train,z_train)
    test_score = model.score(x_test,z_test)
    print( 'Training Score', train_score)
    print ('Testing Score', test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt= DecisionTreeClassifier(criterion ='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train, z_train)
```

```python
mscore(dt)
```

```python
zpred_dt = dt.predict(x_test)
print(zpred_dt)
```

```python
eval_model(z_test,zpred_dt)
```

```python
from sklearn.tree import plot_tree
```

```python
cn = ['no','yes']
fn = x_train.columns
print(fn)
print(cn)
```

```python
plot_tree(dt,class_names=cn,filled=True)
plt.show()
```

```python
dt1 = DecisionTreeClassifier(criterion='entropy',max_depth=4,min_samples_split=15)
dt1.fit(x_train,z_train)
```
